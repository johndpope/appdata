```python
import time
from selenium import webdriver
from pprint import pprint
import concurrent
import pandas as pd
import traceback
```

```python
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
# chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument('user-agent="Mozilla/5.0 (Linux; U; Android 4.0.2; en-us; Galaxy Nexus Build/ICL53F) AppleWebKit/534.30 (KHTML, like Gecko) Version/4.0 Mobile Safari/534.30"')
chrome_prefs = {
      'profile.managed_default_content_settings.images': 2,
    }
chrome_options.add_experimental_option('prefs', chrome_prefs)
```

```python
driver = webdriver.Chrome(options=chrome_options)
```

```python
MAIN_PAGE_URL = 'http://app.qq.com/#id=app_cate'
driver.get(MAIN_PAGE_URL);
```

```python
# get url to category page
CATEGORY_URLS = {}
BASE_INNER_PAGE_URL = 'http://app.qq.com/#id=catelist&cateid={}'
for button in driver.find_elements_by_class_name('list-inner'):
    name = button.get_attribute('data-catename')
    appid = button.get_attribute('data-cateid')
    url = BASE_INNER_PAGE_URL.format(appid)
    CATEGORY_URLS[name] = url
    
CATEGORY_URLS
```

```python
INNER_URLS = set()
BASE_INNER_URL = 'http://app.qq.com/#id=detail&appid={}'

def crawl_category_page(driver, name, url):
    driver.get(url)
    time.sleep(0.2)
    
    # load more by scrolling and clicking button if any
    try:
        for i in range(30):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            try:
                driver.find_element_by_css_selector('a.btn-gray.loadmore').click()
            except:
                pass
            time.sleep(0.1)
    except:
        traceback.print_exc()
        pass
    
    cards = driver.find_element_by_class_name('mod-app-list')\
                  .find_elements_by_tag_name('li')
    for card in cards:
        url = BASE_INNER_URL.format(card.get_attribute('data-appid'))
        INNER_URLS.add(url)

for category_page_name in CATEGORY_URLS:
    crawl_category_page(driver, category_page_name, CATEGORY_URLS[category_page_name])

print(len(INNER_URLS))
INNER_URLS
```

```python
OUTFILENAME = 'app_urls.txt'
with open(OUTFILENAME, 'w') as f:
    for url in INNER_URLS:
        f.write(url+'\n')
```
