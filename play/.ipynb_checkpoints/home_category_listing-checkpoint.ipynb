```python
import time
from selenium import webdriver
from pprint import pprint
import concurrent
import json
import pandas as pd
import random
```

```python
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--disable-extensions")
# we dont need to load images
chrome_prefs = {
      'profile.managed_default_content_settings.images': 2,
    }
chrome_options.add_experimental_option('prefs', chrome_prefs)
def get_driver():
    return webdriver.Chrome(options=chrome_options)
```

```python
driver = get_driver()
driver.get('https://play.google.com/store/apps?hl=en_US')
```

```python
WAIT_TIME_BETWEEN_SCROLL = .5
def scrolltobottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
def scrolltotop(driver):
    driver.execute_script("window.scrollTo(0, 0)")
def scroll(driver, scroll_count):
    for i in range(scroll_count):
        time.sleep(WAIT_TIME_BETWEEN_SCROLL)
        scrolltotop(driver)
        scrolltobottom(driver)
```

```python
def pkg_from_url(url):
    return url.replace('https://play.google.com/store/apps/details?id=','')

# card should be one defined by 'Vpfmgd'
def parse_card(card):
    j = {}
    text_button = card.find_element_by_css_selector('div.b8cIId.ReQCgd.Q9MA7b')
    j['name'] = text_button.text
    url = text_button.find_element_by_tag_name('a').get_attribute('href')
    j['pkg'] = pkg_from_url(url)
    return j
```

```python
parse_card(driver.find_element_by_class_name('Vpfmgd'))
```

```python
def get_outer_category_listing(driver):
#     scroll(driver, 10)
    categories = []
    
    l = driver.find_elements_by_class_name('Ktdaqe  ')
    for box in l:
        c = {}
        
        # name of the category
        c['category_name'] = box.find_element_by_class_name('xwY9Zc').text
        if c['category_name'] == '': continue
        
        # "see more" button if there is one
        try:
            b = box.find_element_by_class_name('W9yFB')
            a = b.find_element_by_tag_name('a')
            c['url'] = a.get_attribute('href')
        except:
            # if there is no see more button, directly parse cards 
            c['apps'] = []
            cards = box.find_elements_by_class_name('Vpfmgd')
            for card in cards:
                c['apps'].append(parse_card(card))
        
        categories.append(c)
    return categories
```

```python
get_outer_category_listing(driver)
```

```python
def crawl_inner_page(driver, url):
    driver.get(url)
#     scroll(driver, 10)
    
    title = driver.find_element_by_css_selector('h2.sv0AUd.bs3Xnd')
    # print(title.text)
    
    # cards are 'wXUyZd' -> <a> has a url, extract pkg 
    cards = driver.find_elements_by_class_name('wXUyZd')
    l = []
    for card in cards:
        a = card.find_element_by_tag_name('a')
        pkg = pkg_from_url(a.get_attribute('href'))
        l.append(pkg)
    return l
```

```python
def crawl_outer_page(driver, url, pagename):
    driver.get(url)
    
    d = dict()
    d['name'] = pagename
    d['url'] = url
    d['content'] = get_outer_category_listing(driver)
    
    for box in d['content']:
        if 'url' not in box: continue
        driver.get(box['url'])
        box['content'] = get_outer_category_listing(driver)
        if len(box['content']) == 1:
            box['content'] = box['content'][0]['apps']
    
    return d
```

```python
d = crawl_outer_page(driver, 
                     'https://play.google.com/store/apps?hl=en_US',
                     'Home')
pprint(d)
```
