```python
import pandas as pd
import json
from datetime import datetime
import random
```

```python
D = {}
with open('data.jl',encoding='utf-8') as f:
    for l in f.readlines():
        try:
            j = json.loads(l)
            pkgname = j['pkg_name']
            D[pkgname] = j
        except:
            pass
len(D)
```

```python
D['com.facebook.orca']
```

```python
def to_int(s):
    if s == '':
        return 0
    if isinstance(s, int):
        return s
    return int(s.replace(',','').replace('+',''))

def clean(d):
    r = {
        'pkgname': d['pkg_name'],
        'appname': d['appname'],
        'numrating': to_int(d['num_rating']),
        'overallrating': float(d['rating']),
        'rating5': to_int(d['rating5']),
        'rating4': to_int(d['rating4']),
        'rating3': to_int(d['rating3']),
        'rating2': to_int(d['rating2']),
        'rating1': to_int(d['rating1']),
        'publisher': d['publisher'],
        'genre': d['genre'],
        'editorschoice': d['is_editors_choice'],
        'screenshots': d['num_screenshot'],
        'mininstallcount': 0,
    }
    try: r['mininstallcount'] = to_int(d['install_count_est'])
    except: pass
    
    try:
        t = datetime.strptime(d['update_date'], '%B %d, %Y')
        r['updatetime'] = t.strftime('%Y-%m-%d')
    except:
        r['updatetime'] = 'unknown'
    
    return r

# D['Android.Page']
clean(D['Android.Page'])
```

```python
d = D[random.choice(list(D.keys()))]
print(d)
clean(d)
```

```python
D1 = {}
for k in D:
    D1[k] = clean(D[k])
D1['com.facebook.orca']    
```

```python
df = pd.DataFrame.from_dict(D1, orient='index')
df
```

```python
df.to_csv('data.csv', index=False)
```
