```python
import json
```

```python
# load everythin in data-*.jl in to a list of dict
ITEMS = []
for i in range(0, 8):
    filename = f'data-{i}.jl'
    print('reading', filename)
    with open(filename, encoding='utf-8') as f:
        for l in f.readlines():
            j = json.loads(l)
            ITEMS.append(j)
```

```python
ITEMS[0]
```

```python
# clean up useless fields in items
# reformat list of rating distribution to individual keys
# sanitize text strings

for item in ITEMS:
    if 'snapshotsUrl' in item: 
        item['numSnapshot'] = len(item['snapshotsUrl'])
        del item['snapshotsUrl']
        
    if 'images' in item: 
        item['numImages'] = len(item['images'])
        del item['images']
        
    if 'apkMd5' in item: del item['apkMd5']
        
    if 'appRatingInfo' in item: 
        item['appAverageRating'] = item['appRatingInfo']['averageRating']
        item['appRatingCount'] = item['appRatingInfo']['ratingCount']
        for i in range(1,6):
            item[f'appRating{i}'] = \
                item['appRatingInfo']['ratingDistribution'][str(i)]
        del item['appRatingInfo']

    if 'description' in item:
        s = item['description']
        item['description'] = s.replace('\r',' ').replace('\n',' ')
        
    if 'newFeature' in item:
        s = item['newFeature']
        item['newFeature'] = s.replace('\r',' ').replace('\n',' ')
        
```
